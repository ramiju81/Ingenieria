```python
def figura_a():
    niveles = [1, 4, 6, 8, 6, 4, 1]
    for nivel in niveles:
        print(" " * ((8 - nivel) // 2) + "*" * nivel)

figura_a()
```

```python
def figura_b():
    niveles = [1, 3, 5, 7, 9, 11, 13]
    for nivel in niveles:
        print(" " * ((13 - nivel) // 2) + "*" * nivel)

figura_b()
```
